```python
from Bio import AlignIO
from Bio import Phylo


from Bio.Phylo.TreeConstruction import DistanceCalculator, DistanceTreeConstructor
import matplotlib.pyplot as plt
import matplotlib
```

```python
aln = AlignIO.read("Fasta_baseplate.aln", "clustal")

calculator = DistanceCalculator("identity")
dm = calculator.get_distance(aln)

constructor = DistanceTreeConstructor()# Construct the phlyogenetic tree using UPGMA algorithm
UPGMATree = constructor.upgma(dm)

# Save the tree to a new file 
Phylo.write(UPGMATree, "Tree_baseplate.xml", "phyloxml")
```

**NOTE**: Before the next step it is needed to manually clean the xml files, in order to remove the name of the intermediate branches ("Inner[0-9]{0,3}"). This should be performed for each protein.

```python


# Convert the tree to a different format (optional)
Phylo.convert("Tree_baseplate.xml", "phyloxml", "Tree_baseplate.nex", "nexus")

```

```python

tree = Phylo.read("Tree_baseplate.nex", "nexus")

fig = plt.figure(figsize=(20, 30), dpi=50) # create figure & set the size 
matplotlib.rc('font', size=12)              # fontsize of the leaf and node labels 
matplotlib.rc('xtick', labelsize=20)       # fontsize of the tick labels
matplotlib.rc('ytick', labelsize=20)       # fontsize of the tick labels

#turtle_tree.ladderize()
axes = fig.add_subplot(1, 1, 1)
axes.axvline(x=0.3, color='r', linestyle='--')
Phylo.draw(tree, axes=axes)
fig.savefig("0.3_baseplate_cladogram_2.png")
```

## Number of clusters

**treshold 0.3** = 5 clusters, 1 singleton
